```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import math
import random
from collections import Counter

model = joblib.load("march_madness_model.pkl")
scaler = joblib.load("scaler.pkl")

df = pd.read_csv("cbb26_prepared.csv")

features = [
    "WP", "ADJOE", "ADJDE", "NET_EFF", "BARTHAG", "TOR", "TORD",
    "ORB", "DRB", "FTR", "FTRD", "2P_O", "2P_D", "3P_O", "3P_D", "RK"
]

# remove duplicate teams
df = df.drop_duplicates(subset="TEAM").copy()

X_scaled = scaler.transform(df[features])

probs = model.predict_proba(X_scaled)[:, 1]
probs = np.clip(probs, 1e-6, 1 - 1e-6)

df["MODEL_PROB"] = probs
df["TEAM_STRENGTH"] = np.log(probs / (1 - probs))

ranked = df.sort_values("TEAM_STRENGTH", ascending=False).reset_index(drop=True)
```
